# Problem Statement

The task is to detect fraudulent credit card transactions within a highly imbalanced dataset where fraudulent cases are extremely rare compared to legitimate transactions. The objective is to build a machine learning model that can effectively identify these frauds despite the imbalance, improving detection rates while minimizing false alarms.


```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from collections import Counter
```

# Part A: Baseline Model and Data Analysis
## 1) Data Loading and Analysis

First, I load the credit card transactions dataset to begin the analysis. The dataset contains 284,807 records with 31 features each, including anonymized principal components (V1 to V28), transaction amount, and the class label indicating whether the transaction is fraudulent or valid.

By examining the first few rows, I get an initial sense of the data structure and the types of features available for modeling.

This step lays the groundwork for understanding the dataset and preparing it for further processing and analysis.


```python
# Load dataset
df = pd.read_csv("creditcard.csv")
print(f"Dataset shape: {df.shape}")
df.head()
```

# Class Distribution Visualization

Next, I examine the distribution of the valid and fraudulent transactions in the dataset.

Using a pie chart, I visualize the class imbalance to get a clear picture of how skewed the data is.

The pie chart indicates that the majority of transactions are valid (non-fraud), which far outnumber the fraudulent ones. Specifically, only a tiny fraction of the transactions are labeled as fraud.

This severe imbalance highlights the challenge ahead: models trained without special handling may predominantly predict the majority class, resulting in poor fraud detection performance.

Understanding this imbalance is crucial before applying any modeling or resampling techniques.


```python
counts = df['Class'].value_counts()
plt.figure(figsize=(4, 5))
plt.pie(counts.values, 
        labels=[f"Valid ({counts})", f"Fraud ({counts[1]})"], 
        autopct='%2.3f%%', 
        explode=[0, 0.15], 
        colors=["cyan", "red"])
plt.title("Class Distribution")
plt.legend(title="classes", loc=1, bbox_to_anchor=(1.25, 1), fontsize=8)
plt.show()
```

# 2) Model Training
## Train-Test Split with Stratification

To prepare the data for modeling, I split the dataset into training and testing sets in a 70:30 ratio.

I use stratified sampling to ensure that the class distribution in both training and testing sets preserves the original imbalance present in the full dataset.

This stratification is important because it maintains realistic proportions of fraudulent and valid transactions in both sets, which is critical for fair evaluation of model performance.

After splitting, I print the shapes of the training and testing sets and verify the class distributions to confirm that stratification was successful.


```python
X = df.drop(columns=['Class'])
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    train_size=0.7, random_state=42)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train class distribution:", Counter(y_train))
print("Test class distribution:", Counter(y_test))
```

# Feature Scaling

Before training models, I apply feature scaling to normalize the data.

Specifically, I use `StandardScaler` to transform the features so that they have zero mean and unit variance.

Feature scaling is critical because many machine learning algorithms, including Logistic Regression and clustering methods used later, perform better and converge faster when features are on similar scales.

I fit the scaler on the training data to learn the scaling parameters (mean and standard deviation), and then apply the same transformation to both the training and testing data to ensure consistency.

Finally, I convert the scaled feature arrays back into Pandas DataFrames, preserving the original feature names and indices for interpretability throughout the modeling process.


```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns, index=X_test.index)
```

## Baseline Logistic Regression Model

I build a baseline classification model using Logistic Regression to understand the initial performance on the imbalanced credit card fraud dataset.

The model is trained on the scaled training data, with `class_weight='balanced'` to give higher importance to the minority (fraud) class during training, which helps mitigate imbalance bias.

After training, I evaluate the model’s performance on the scaled test data.

I report the overall accuracy, but given the class imbalance, this metric alone may be misleading.

Therefore, I also analyze the confusion matrix and detailed classification report with precision, recall, and F1-score for both valid and fraud classes.

These metrics provide a more comprehensive view of the model’s effectiveness in detecting fraudulent transactions, which is the primary objective.


```python
model1 = LogisticRegression(solver="lbfgs", max_iter=1000, class_weight="balanced")
model1.fit(X_train_scaled, y_train)
y_pred = model1.predict(X_test_scaled)
print("Logistic Regression Accuracy:", model1.score(X_test_scaled, y_test))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=["Valid", "Fraud"]))
```

# Baseline Evaluation: 

For evaluating the model's performance on this imbalanced fraud detection problem, metrics such as **Precision, Recall, and F1-score** for the minority class are more informative than accuracy because:

- **Accuracy can be misleading in imbalanced datasets** like credit card fraud detection where the vast majority of transactions are valid and only a tiny fraction are fraudulent. For example, predicting "valid" for every transaction can yield high accuracy (over 99%) but completely fail to identify fraud.

- **Precision** tells me, out of all transactions predicted as fraud, how many were actually fraud. A low precision means many false alarms, which can be costly and cause unnecessary manual checks.

- **Recall** indicates how many of the actual fraudulent transactions my model successfully detected. High recall is critical to minimize missed fraud cases, which directly impacts financial loss.

- **F1-score** balances precision and recall by computing their harmonic mean. This single metric gives a more holistic view of minority class performance, especially when one metric alone (precision or recall) might be misleading.

In fraud detection, it’s often more important to optimize for recall (catch maximum frauds) while keeping precision reasonably high to avoid too many false positives—making precision, recall, and F1-score far more valuable for evaluation than accuracy.

This understanding guides me to focus on improving and interpreting these metrics carefully throughout the modeling process.


```python
plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred),
            annot=True, fmt="d", cmap="coolwarm",
            xticklabels=["Valid", "Fraud"], yticklabels=["Valid", "Fraud"])
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Part B: Gaussian Mixture Model (GMM) for Synthetic Sampling
# 1) Theoretical Foundation

# Difference Between GMM-Based Synthetic Sampling and SMOTE

**SMOTE (Synthetic Minority Oversampling Technique)** generates synthetic samples by interpolating linearly between existing minority class samples and their nearest neighbors. It assumes that the minority class forms a single, homogeneous cluster and creates synthetic points along the lines connecting nearby samples. While simple and effective in many cases, SMOTE may struggle when the minority class distribution is complex, multimodal, or composed of multiple distinct subgroups.

**GMM-Based Synthetic Sampling**, on the other hand, uses a probabilistic model (Gaussian Mixture Model) to learn the underlying distribution of the minority class as a mixture of multiple Gaussian components. This allows it to capture complex and multimodal structures within the minority data.

# Why GMM is Theoretically Better

- **Capturing Sub-Groups:** GMM models the minority class as a combination of several Gaussian distributions (subgroups), enabling it to generate synthetic samples that respect the multi-cluster nature of the data.

- **Modeling Complex Shapes:** By fitting full covariance matrices, GMM can capture elliptical data shapes and correlations between features, which linear interpolation methods like SMOTE do not consider.

- **Reducing Over-Simplification:** SMOTE’s linear interpolation can create synthetic samples in non-representative or even noisy regions between clusters, whereas GMM sampling generates samples according to the learned probability density, leading to more realistic and representative synthetic points.

Therefore, GMM-based synthetic sampling is theoretically better at modeling diverse minority class structures, leading to synthetic samples that more accurately represent the true data distribution, especially when the minority class exhibits multiple modes or complex shapes.


## 2) Determining the Optimal Number of Components (k) for GMM

To select the optimal number of Gaussian components \( k \) for the Gaussian Mixture Model (GMM), I employed model selection criteria that balance model fit and complexity:

- **Akaike Information Criterion (AIC)** and **Bayesian Information Criterion (BIC)** are statistical metrics that measure how well a model fits the data while penalizing for the number of parameters to avoid overfitting.

- I trained multiple GMMs with different values of \( k \) ranging from 1 up to 14 on the minority class data.

- For each model, I calculated the AIC and BIC scores; lower values indicate a better trade-off between fit and simplicity.

- By plotting these scores against \( k \), I identified the point where the scores reached their minimum or leveled off, indicating the optimal number of components that capture the data's structure without overfitting.

- In this analysis, \( k=6 \) was selected as it showed the lowest AIC and BIC values, suggesting that four Gaussian components best represent the underlying minority class distribution.

This systematic approach ensures the model complexity is appropriate and aids in generating realistic synthetic samples from the GMM.


```python
from sklearn.mixture import GaussianMixture

X_minority = X_train_scaled[y_train == 1]
N_COMPONENTS = range(1, 15)
aic_vals, bic_vals = [], []

for k in N_COMPONENTS:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(X_minority)
    aic_vals.append(gmm.aic(X_minority))
    bic_vals.append(gmm.bic(X_minority))

plt.figure(figsize=(8, 5))
plt.plot(N_COMPONENTS, aic_vals, '-o', label='AIC')
plt.plot(N_COMPONENTS, bic_vals, '-s', label='BIC')
plt.xlabel("Number of Gaussian Components")
plt.ylabel("Score (lower is better)")
plt.title("Model Selection using AIC & BIC")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

I have chosen to use \( k=6 \) Gaussian components for the GMM synthetic sampling after carefully considering the options \( k=4 \), \( k=5 \), and \( k=6 \).

The main reason for this choice is that \( k=6 \) represents a clear **elbow point** in both the AIC and BIC curves where the improvement in model fit slows down significantly.

Selecting \( k=6 \) strikes a good balance between capturing sufficient minority class structure and maintaining model simplicity, helping to avoid overfitting.

Moreover, a less complex model is generally more robust when generating synthetic samples, ensuring that the synthetic data remains representative and does not overfit noise or outliers.

Therefore, \( k=6 \) is a practical, well-justified choice that balances model accuracy and generalizability.


##  3) Synthetic Minority Sample Generation using Gaussian Mixture Model (GMM)

To better balance the training dataset and improve fraud detection, I use a Gaussian Mixture Model (GMM) to generate synthetic minority class samples that resemble the distribution of actual fraudulent transactions.

I first fit a GMM with 4 components (chosen based on model selection criteria) to the scaled minority class data.

Next, I calculate the number of synthetic samples needed to balance the minority class to the majority class size by subtracting the number of original minority samples from the majority samples.

Using the fitted GMM, I generate the required number of synthetic minority samples by sampling from the learned Gaussian mixture distribution.

To integrate these synthetic samples seamlessly with the original training data, I convert them into a DataFrame with the same column names as the scaled training features.

I then combine the original majority class samples, the original minority class samples, and the synthetic minority samples into a single balanced training dataset.

Corresponding class labels are created to reflect the majority and the combined minority classes.

Finally, I print the class distribution of the augmented training set to confirm successful balancing.

This approach aims to provide a more representative and diversified minority class to the classifier, which should improve its ability to detect fraudulent transactions.


```python
# Fit GMM to minority class
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=8)
gmm.fit(X_minority)

# Calculate number of synthetic samples needed (using number of rows)
n_synthetic = X_train_scaled[y_train == 0].shape[0] - X_minority.shape[0]

# Generate synthetic minority samples
X_synthetic, _ = gmm.sample(n_samples=n_synthetic)

# Convert synthetic samples to DataFrame with proper column names
X_synthetic_df = pd.DataFrame(X_synthetic, columns=X_train_scaled.columns)

# Create combined training set with original majority, original minority, and synthetic minority data
X_train_gmm = pd.concat([
    X_train_scaled[y_train == 0],
    X_minority,
    X_synthetic_df
], ignore_index=True)

# Create combined label array
y_train_gmm = np.hstack([
    np.zeros(X_train_scaled[y_train == 0].shape[0]),   # majority class labels
    np.ones(X_minority.shape[0]),                       # original minority labels
    np.ones(n_synthetic)                                # synthetic minority labels
])

# Print class distribution after synthetic sampling
from collections import Counter
print("Class distribution after synthetic sampling:", Counter(y_train_gmm.astype(int)))
```

# Cluster-Based Undersampling (CBU) of Majority Class

To address the overwhelming size of the majority class, I apply a cluster-based undersampling method to reduce its dominance while preserving its internal structure.

I start by clustering the majority class samples using KMeans clustering.

Using the Elbow method, I determine the optimal number of clusters by plotting the within-cluster sum of squares (WCSS or inertia) for different counts of clusters and identifying the point where the rate of decrease sharply changes.

Suppose the optimal number of clusters is 3, I then fit KMeans with 3 clusters and assign each majority sample to a cluster.

Next, I calculate the target size for the majority class to match the combined size of the minority class and the synthetic samples generated earlier.

I undersample each cluster evenly: if a cluster contains more samples than the target share, I randomly resample without replacement down to the target; otherwise, I keep all samples.

After undersampling all clusters, I combine the reduced majority clusters to form a balanced majority set.

Finally, I combine this undersampled majority set with the original minority and synthetic minority samples.

This balanced dataset now better represents both classes, reducing bias and helping classifiers learn minority patterns more effectively.

I print the class distribution after undersampling to verify the balance achieved.


```python
X_majority = X_train_scaled[y_train == 0]

# Determine optimal clusters for KMeans (Elbow method)
inertia = []
K_range = range(1, 16)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=8, n_init=10)
    kmeans.fit(X_majority)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(K_range, inertia, marker="s", color="purple")
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS (inertia)")
plt.title("KMeans Elbow Curve (Majority Class)")
plt.grid(alpha=0.3)
plt.show()

K = 3  # Suppose optimal clusters = 3
kmeans = KMeans(n_clusters=K, random_state=8)
clusters = kmeans.fit_predict(X_majority)

X_majority_df = pd.DataFrame(X_majority)
X_majority_df['cluster'] = clusters

# Calculate target size: total number of minority samples (original + synthetic)
target_size = X_minority.shape[0] + X_synthetic.shape[0]

# Undersample majority class from each cluster evenly
n_samples_per_cluster = target_size // K

X_majority_balanced = []
for c in range(K):
    cluster_data = X_majority_df[X_majority_df['cluster'] == c].drop('cluster', axis=1).values
    
    # Undersample if cluster size is larger than desired samples per cluster
    if len(cluster_data) > n_samples_per_cluster:
        cluster_balanced = resample(cluster_data, replace=False, n_samples=n_samples_per_cluster, random_state=8)
    else:
        cluster_balanced = cluster_data  # Use all if smaller
    
    X_majority_balanced.append(cluster_balanced)

# Combine all undersampled cluster data
X_majority_balanced = np.vstack(X_majority_balanced)
y_majority_balanced = np.zeros(X_majority_balanced.shape[0])  # Labels for majority class

# Combine undersampled majority with minority (original + synthetic)
X_balanced = np.vstack([X_majority_balanced, X_minority, X_synthetic])
y_balanced = np.hstack([y_majority_balanced,
                        np.ones(X_minority.shape[0]),
                        np.ones(X_synthetic.shape[0])])

print("Class distribution after CBU:", np.bincount(y_balanced.astype(int)))
```
